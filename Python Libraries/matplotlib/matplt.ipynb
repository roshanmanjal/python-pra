```python
# multivariate analysis
df["Exp"] = [1,2,3,4,5,6,7,8]
```

```python
df
```

```python
# bubble plot
plt.scatter(df["Age"], df["Salary"], s = df["Exp"] * 15, color = "green", edgecolor = "red")  # 3 njumerical columns
plt.title("Age vs Salary vs Exp")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```

```python
# 2 numeric data and 1 category data column
# dept_codes = df["dept"].astype("category").cat.codes
plt.scatter(
    df["Age"],
    df["Salary"],
    c=df["dept"].map({"HR" :"yellow", "Assistant" : "black", "Manager" : "Green", "Intern" : "Red", "Trainee" : "skyblue" })
)
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Age vs Salary by Dept")
plt.colorbar(label="Department Code")
plt.show()
```

```python
# to also print legend gen view so add color to label it
color = {"HR": "yellow","Assistant" : "black", "Intern":"red", "Manager": "green", "Trainee" :"grey" }

for dept, color in color.items():
    df_dept = df[df["dept"] == dept]
    plt.scatter(df_dept["Age"], df_dept["Salary"], c = color, label = dept)
    plt.legend()  # In this plt.show will not come as it will print seperate to see in 1 table only plt.legend will show
```

```python
# object oriented API means sub plot in single box
fig,axs = plt.subplots(1,3, figsize = (15,5))

# line plot = 1st box
axs[0].plot(df["Age"], df["Salary"], color = "red", marker = "*", linewidth = "2", markersize = 2)
axs[0].grid()
axs[0].set_xlabel("Age")
axs[0].set_ylabel("Salary")
axs[0].set_title("Line PLot")

# histogram plot i.e 2nd box
axs[1].hist(df["Salary"], bins = 7, color = "skyblue")
axs[1].grid()
axs[1].set_xlabel("Salary")
axs[1].set_ylabel("Age")
axs[1].set_title("Histogram")

# Box plot 3rd box
axs[2].boxplot(df["Salary"])
axs[2].set_title("Boxplot")
axs[2].set_ylabel("Salary")
plt.show()
```

```python
# multiple integers in one plot
Data = {
    "Sales" : [100,200,250,300],
    "Year" : [2010, 2020, 2025,2026],
    "Profit" : [20,30,40,50],
    "Expenses" : [12,14,15,16]
}
df2 = pd.DataFrame(Data)
```

```python
plt.plot(df2["Year"], df2["Sales"], label = "Sales")
plt.plot(df2["Year"], df2["Profit"], label = "Profit")
plt.plot(df2["Year"], df2["Expenses"], label = "Expenses")
plt.legend()
plt.xlabel("Year")
plt.ylabel("Amount")
plt.title("Financial Analysis")
plt.show()
```
